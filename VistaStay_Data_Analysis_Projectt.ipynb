### ==> 3. Data Transformation
***

**Create occupancy percentage column**

```python
df_agg_bookings.head(3)
```

```python
df_agg_bookings['occ_pct'] = df_agg_bookings.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
```

You can use following approach to get rid of SettingWithCopyWarning

```python
new_col = df_agg_bookings.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
df_agg_bookings = df_agg_bookings.assign(occ_pct=new_col.values)
df_agg_bookings.head(3)
```

Convert it to a percentage value

```python
df_agg_bookings['occ_pct'] = df_agg_bookings['occ_pct'].apply(lambda x: round(x*100, 2))
df_agg_bookings.head(3)
```

**2. Print average occupancy rate per city**

```python
df_hotels.head(3)
```

```python
df = pd.merge(df, df_hotels, on="property_id")
df.head(3)
```

```python
avg_occ_city = df.groupby("city")["occ_pct"].mean().sort_values(ascending=False)
avg_occ_city
```

```python
plt.figure(figsize=(8, 5))

sns.barplot(x=avg_occ_city.index, y=avg_occ_city.values, hue=avg_occ_city.index, palette='coolwarm', legend=False)

for i, v in enumerate(avg_occ_city.values):
    plt.text(i, v + 0.8, f"{v:.2f}%", ha='center', fontweight='bold', color='black')

plt.title("Average Occupancy Rate by City", fontsize=14, fontweight='bold')
plt.xlabel("City", fontsize=12)
plt.ylabel("Average Occupancy (%)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**Insight:**  
Delhi shows the highest average occupancy rate among all cities, followed by Hyderabad and Mumbai.  
This indicates that hotels in Delhi have better room utilization compared to other locations.


**3. When was the occupancy better? Weekday or Weekend?**

```python
df_date.head(3)
```

```python
df = pd.merge(df, df_date, left_on="check_in_date", right_on="date")
df.head(3)
```

```python
# Calculate average occupancy by day type
avg_occ_day = df.groupby("day_type")["occ_pct"].mean().round(2).reset_index()

# Plot
plt.figure(figsize=(6,4))
sns.barplot(
    data=avg_occ_day, 
    x="day_type", 
    y="occ_pct", 
    hue="day_type",
    palette="viridis", 
    legend=False       
)

# Add labels on bars
for i, v in enumerate(avg_occ_day["occ_pct"]):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', fontweight='bold', color='black')

# Customize chart
plt.title("Average Occupancy Rate: Weekday vs Weekend", fontsize=14, fontweight='bold')
plt.xlabel("Day Type", fontsize=12)
plt.ylabel("Average Occupancy (%)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**Insight:**  
Occupancy rates are noticeably higher on weekends (`72%`) compared to weekdays (`51%`).  
This indicates that customer bookings increase during weekends, likely due to leisure and travel activities.


**4: In the month of June, what is the occupancy for different cities**

```python
df_june_22 = df[df["mmm yy"]=="Jun 22"]
df_june_22.head(4)
```

```python
df_june_22.groupby('city')['occ_pct'].mean().round(2).sort_values(ascending=False)
```

```python
# Group and sort by average occupancy in June
avg_occ_june = (
    df_june_22.groupby('city')['occ_pct']
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index()
)

# Plot the data
plt.figure(figsize=(7,5))
sns.barplot(
    data=avg_occ_june,
    x='city',
    y='occ_pct',
    hue='city',         
    palette='crest',
    legend=False
)

# Add labels on top of bars
for i, v in enumerate(avg_occ_june['occ_pct']):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', fontweight='bold', color='black')

# Chart titles and styling
plt.title("Average Occupancy Rate by City (June 2022)", fontsize=14, fontweight='bold')
plt.xlabel("City", fontsize=12)
plt.ylabel("Average Occupancy (%)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**Insight:**  
In June 2022, **Delhi** recorded the highest average occupancy rate (~62.47%), 
followed by **Hyderabad** and **Mumbai**, while **Bangalore** showed the lowest.  
This indicates that Delhi hotels experienced better utilization during this month.
